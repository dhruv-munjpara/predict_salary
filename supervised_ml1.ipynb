```python
import pandas as pd
import matplotlib.pyplot as plt
from  sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from  sklearn.metrics import r2_score
df=pd.read_csv("Salary_dataset.csv")
df
```

```python
df.columns
```

```python
x=df[['YearsExperience']]
y=df['Salary']

print("x head:\n",x.head())
print("y head:\n",y.head())
```

```python
# data split -x train x test y train y test -----train test split
# test_size=0.20//20 80 trainging random_state=42//integer fun/reference

x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.20,random_state=42
)
```

```python
model=LinearRegression()  #model linear constructor same linear class innistialization
```

```python
model.fit(x_train,y_train)   #input and output
```

```python
v=model.predict(x_test)
score=r2_score(v,y_test)
print(score)
```

```python
#y=mx+c
#output=slope(input)+constant
print(model.coef_[0])
print(model.intercept_)
```

```python
result=model.predict([[8]])
result
```

```python
plt.scatter(x,y,color="blue")
plt.plot(x,model.predict(x),color="red")
plt.xlabel("exp")
plt.ylabel("salary")
plt.title("exp vs salry")
plt.show()
```
