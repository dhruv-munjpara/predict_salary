```python
import pandas as pd
import matplotlib.pyplot as plt
from  sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

df=pd.read_csv("Salary_data.csv")
df
```

```python
df.columns
```

```python
x=df[["YearsExperience","Age","Education"]]
y=df["Salary"]

print("x head:\n",x.head())
print("y head:\n",y.head())
```

```python
x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.20,random_state=42
)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
v=model.predict(x_test)
score=r2_score(v,y_test)
print(score)
```

```python
print(model.coef_[0])
print(model.intercept_)
```

```python
result=model.predict([[9,22,15]])
print(result)
```

```python
plt.scatter(x['YearsExperience'],y,color="blue")
plt.plot(x['YearsExperience'],model.predict(x),color="red")
plt.title("year_age_education wise salary")
plt.xlabel("y_a_e")
plt.ylabel("salary")
plt.show()
```
